##**INTRODUCTION**

Transition metal dichalcogenides (TMDCs) provide a promising alternative to silicon and germanium semiconductor applications as well as other common materials.<sup>1</sup> TMDCs have the chemical formula MX2, where M is a transition element and X is chalcogen.<sup>1</sup> Typically the bandgap of TMDCs is typically in the range of visible light, which makes them candidates for use in photovoltaic cells.<sup>1</sup> Conversely TMDCs have potential to produce flexible and transparent LEDs, although the required doping makes fabrication difficult.<sup>1</sup>

###Semiconductor Basics
* Semiconductors typically act as insulators
* Electrons can be moved to bands where mobility allows for conduction
* High mobility allows for faster transistor switching

###Recombination and Carriers
* Band-to-band: an electron from the conduction band (E<sub>c</sub>)  combines with a hole in the valence band (E<sub>v</sub>) and releases energy in the form of a photon
* Trap-assisted: an electron from the conduction band falls into an energy level within the bandgap caused by the presence of a foreign atom or a structural defect before moving into the the valence band
* Auger: an electron and a hole recombine in a band-to-band transition and the resulting energy is given off to another electron or hole

###Methods/Actions
####Photoluminescence Spectroscopy
Photons from a laser (light source) hit the nanomaterial. The electrons in the material are excited and can jump to an excited state (higher energy level). When electrons drop from the conduction band to the valence band and energy is released in the form of a luminescence photon.

####Laboratory Setup
We collected photoluminescence data by directing a laser onto the center of the nanomaterial layer, and then recording the intensity of the emitted photon at a certain location for the entire lifetime of the process. The location from which the emission is measured is moved and a laser pulse is directed to the center again, and another emission is measured over the lifetime of the process. The data we collected represents the intensity at every pixel in the material.

##Conclusion
* Recombination: Band-to-band recombination has a 34.3 times larger effect than trap-assisted recombination initially,  but decreases to 3.11 times larger
* Diffusion: The rate of diffusion decreases over time as carrier density decreases. D<sub>1</sub> = 0.74 cm<sup>2</sup> s<sup>-1</sup> over first ~400 ps, D<sub>2</sub>  = 0.24 cm<sup>2</sup> s<sup>-1</sup> over next ~400ps
